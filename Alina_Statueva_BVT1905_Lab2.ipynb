```python
# Задание 1 (поиски)

# Бинарный поиск
#Генерация нашего массива
from random import random
N = int(input("Длина массива: "))
array = []
for i in range(N):
    array.append(int(random()*100))
array.sort()
print("Наш массив:",array)

number = int(input("Введите искомое значение: "))

low = 0#граница минимальная
high = N-1#граница максимальная
while low <= high:
    mid = (low + high) // 2 #делим пополам
    if number < array[mid]:
        high = mid - 1
    elif number > array[mid]:
        low = mid + 1
    else:
        print("ID нашего значения = ", mid)
        break
else:
    print("Такого числа в нашем массиве нету!")
```

```python
# Бинарное дерево
import time
start = time.time()
class Node:

    def __init__(self, data):

        self.left = None
        self.right = None
        self.data = data

#Метод создания нашего дерева
    def insert(self, data):

        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data
            
#метод сравнения и поиска нашего числа
    def findval(self, lkpval):
        if lkpval < self.data:# если наше значение меньше, передвигаемся по этой ветке (влево)
            if self.left is None:# если мы долшли до конца и значения нет, то вывод "не найдено"
                return "Значение не найдено"
            return self.left.findval(lkpval)
        elif lkpval > self.data:# если наше значение больше, передвигаемся по этой ветке(вправо)
            if self.right is None:# если мы долшли до конца и значения нет, то вывод "не найдено"
                return "Значение не найдено"
            return self.right.findval(lkpval)
        else:# алгоритм привел нас к значению, следовательно выводим его
            print('Искомое значение найдено ')


number = int(input("Введите искомое значение: "))
#создание дерева
root = Node(35)
root.insert(20)
root.insert(15)
root.insert(4)
root.insert(12)
root.insert(8)
root.insert(44)
root.insert(8)
root.insert(9)
print(root.findval(number))
print("Время поиска: ", (time.time() - start)*1000)


```

```python
# Фибоначчиев поиск
import time
from random import random
N = int(input("Длина массива: "))
array = []
for i in range(N):
    array.append(int(random()*100))
array.sort()
print("Наш массив: ",array)
x = int(input("Введите искомое значение: "))

start = time.time()
def FibonacciSearch(lys, val):
    
    # Инициализируем числа Фибоначчи
    fibM_minus_2 = 0
    fibM_minus_1 = 1
    fibM = fibM_minus_1 + fibM_minus_2
    
    # fibM будет хранить самые маленькие
    # Число Фибоначчи больше или равно n
    while (fibM < len(lys)):
        fibM_minus_2 = fibM_minus_1
        fibM_minus_1 = fibM
        fibM = fibM_minus_1 + fibM_minus_2
    index = -1;# Отмечает удаленный диапазон спереди
    
    # пока есть элементы, которые нужно проверить.
    # Обратите внимание, что мы сравниваем arr [fibM_minus_2] с x.
    # Когда fibM становится 1, fibM_minus_2 становится 0
    while (fibM > 1):
        
        # Проверяем, является ли fibMm2 допустимым местоположением
        i = min(index + fibM_minus_2, (len(lys)-1))
        
        # Если x больше значения в
        # index fibMm2, вырезать подмассив
        # от смещения до i
        if (lys[i] < val):
            fibM = fibM_minus_1
            fibM_minus_1 = fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
            index = i
            
        # Если x меньше значения в
        # index fibMm2, вырезать подмассив
        # после i + 1
        elif (lys[i] > val):
            fibM = fibM_minus_2
            fibM_minus_1 = fibM_minus_1 - fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
        #Обнаружен элемент. индекс возврата
        else :
            return i+1
    
    # сравнение последнего элемента с x * /
    if(fibM_minus_1 and index < (len(lys)-1) and lys[index+1] == val):
        return index+1;
    # элемент не найден. возврат -1
    return "отсутсвует"
print("Индекс искомого значения: ", FibonacciSearch(array, x))
print("Время поиска: ", (time.time() - start)*1000)


```

```python
# Интерполяционный поиск
from random import random
N = int(input("Длина массива: "))
array = []
for i in range(N):
    array.append(int(random()*100))
array.sort()
print("Наш массив: ",array)
x = int(input("Введите искомое значение: "))

start = time.time()
def InterpolationSearch(lys, val):
    low = 0
    high = (len(lys) - 1)
    while low <= high and val >= lys[low] and val <= lys[high]:
        index = low + int(((float(high - low) / ( lys[high] - lys[low])) * ( val - lys[low])))
        if lys[index] == val:
            return index
        if lys[index] < val:
            low = index + 1;
        else:
            high = index - 1;
    return "отсутсвует"

print("Индекс искомого значения: ", InterpolationSearch(array, x))
print("Время поиска: ", (time.time() - start)*1000)
```

```python
# Задание 2 (хеширование)
def hash(astring, tablesize):# функция хиширования
    sum = 0
    for pos in range(len(astring)):
        sum = sum + ord(astring[pos])# возврат значения символа и его сумма с предыдушим

    return sum%tablesize #сумма индексов наших символов

print("Результат: ",hash(["a","c","t"], 11))# 312 % 11 = 4
```

```python
# Рехеширование
from random import random
N = int(input("Длина массива: "))
array = []
for i in range(N):
    array.append(int(random()*100))
array.sort()
print("Наш массив: ",array)

def rehash(oldhash, table_size):
    return (oldhash+1) % table_size

def lp_hash(item_list, table_size):
    
    lp_hash_table = dict([(i,None) for i,x in enumerate(range(table_size))])

    for item in item_list:
        i = item % table_size
        print("%s хеш-функция == %s \n" %(item, i))
        if lp_hash_table[i] == None:
            lp_hash_table[i] = item
        elif lp_hash_table[i] != None:
            print("Колизия \n")
            next_slot = rehash(i, table_size)
            print("Устанавливаем следующий слот %s \n" % next_slot)
            while lp_hash_table[next_slot] != None:
                next_slot = rehash(next_slot, len(lp_hash_table.keys()))
                print("Следующий слот занят, попробуем следующий слот %s \n" % next_slot)
            if lp_hash_table[next_slot] == None:
                lp_hash_table[next_slot] = item
    return lp_hash_table

print(lp_hash(array, 11))
```
